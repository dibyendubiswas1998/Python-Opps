## **Example-04:**

```python
class Persons:

  def __init__(self, fname, lname, age=23):
    self.fname = fname
    self.lname = lname
    self.age = age

  def Prniting(self):
    print(f"Name: {self.fname} {self.lname} & Age: {self.age}")

  def __str__(self):
    return f"Name: {self.fname} {self.lname} & Age: {self.age}"

obj1 = Persons('Dibyendu', 'Biswas')
obj2 = Persons('Arko', 'Paul', 24)
obj3 = Persons('Ram', 'Sharma', 33)
obj4 = Persons('Rocky', 'Sen')


print(obj1)
print(obj2)
print(obj3)
print(obj4)
```

## **Example-05:**

```python
class MyClass:

  def __init__(self, fname, lname):
    self.fname = fname
    self.lname = lname
  
  def Age(self, age):
    self.age = age
    print("My age is ", self.age)
  
  def Location(self, loc):
    self.location = loc
    print(f"I am from {self.location}")
  
  def __str__(self):
    return f"I am {self.fname} {self.lname}"


obj1 = MyClass('Dibyendu', 'Biswas')
obj1.Age(23)
obj1.Location('Siliguri')
print(obj1)
```

## **Example-07:**

```python
class ExampleClass:

  def __init__(self, fname, lname):
    self.fname = fname
    self.lname = lname
  
  def Age(self, age):
    self.age = age
    return self.age
  
  def Location(self, loc):
    return loc
  
  def PinCode(self, pin):
    return pin

  def Print(self):
    self.age = self.Age(23)
    self.location = self.Location('Siliguri')
    pin = self.PinCode(734001)
    return "My age {age} and location is {location} and pincode {pincode}".format(age=self.age, location=self.location, pincode=pin)

  def __str__(self):
    self.age = self.Age(23)
    self.location = self.Location('Siliguri')
    self.pin = self.PinCode(734001)
    return f"I'm {self.fname} {self.lname}, my age is {self.age}, I'm from {self.location}, pincode {self.pin}"

obj1 = ExampleClass('Dibyendu', 'Biswas')
print(obj1.Print())
print(obj1)
```

## **Example-08:**

```python
def Num():
  return 10

class Call:

  def __init__(self, fname, lname):
    self.num = Num()
    self.fname = fname
    self.lname = lname
  
  def PrintAll(self):
    print(f"Name: {self.fname} {self.lname} and random number: {self.num}")
  
obj = Call('Dibyendu', 'Biswas')
obj.PrintAll()
```

## **Example-09:**

```python
def Add(a, b):
  return a+b

class Calling:

  def __init__(self, name, num1, num2):
    self.name = name
    self.num1 = num1
    self.num2 = num2
    self.add = Add(self.num1, self.num2)

  def PrintAll(self):
    print(f"name: {self.name} & Addition value: {self.add}")


obj = Calling('Dibyendu', 20, 30)
obj.PrintAll()
```

## **Example-10:**

```python
class Person:

    def set_name(self, name):
        self.name = name

    def set_surname(self, surname):
        self.surname = surname
         
    def set_year_of_birth(self, year_birth):
        self.year_birth = year_birth
     
    def __str__(self):
        return f"{self.name} {self.surname} was born in {self.year_birth}"
    

obj = Person()
obj.set_name("Dibyendu")
obj.set_surname("Biswas")
obj.set_year_of_birth('12-03-1999')
print(obj)
```
