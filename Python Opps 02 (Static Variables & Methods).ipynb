# **1. Static Variables:**


"When we declare a variable inside a class, but outside the method, it is called a static or class variable. It can be called directly from a class but not through the instances of a class."

## **Example-01:**

```python
class Employee:
  
  department_name = 'Data Science'   # static variable
  
  def __init__(self, fname, lname, age):
    self.fname = fname
    self.lname = lname
    self.age = age
  
obj = Employee('Dibyendu', 'Biswas', 23)
print(obj.fname)
print(obj.lname)
print(obj.age)
print(obj.department_name)
```

```python
obj1 = Employee('Ram', 'Kumar', 25)
obj1.department_name = 'Development'

obj2 = Employee('Pritam', 'Sen', 34)
obj2.department_name = 'Networking'


print('For Object 1:---------')
print(f"Name: {obj1.fname} {obj1.lname},  Department Name: {obj1.department_name} \n")

print('For Object 2:---------')
print(f"Name: {obj2.fname} {obj2.lname},  Department Name: {obj2.department_name} \n")

print(Employee.department_name)
```

## **Example-02:**

```python
class NewCar:
  car_name = 'Hyundai Verna'      # static variable
  car_type = 'Petrol Automatic'   # static variable
  car_varient = 'SX IVT Opt'      # static variable

car1 = NewCar()
print(f"{car1.car_name}, {car1.car_type}, {car1.car_varient} \n")

# Declare the other car deails
car2 = NewCar()
car2.car_name = "Mahindra Thar"
car2.car_type = 'Petrol Automatic'
car2.car_varient = 'LX 4-Str Hard Top AT'
print(f"{car2.car_name}, {car2.car_type}, {car2.car_varient} \n")
```

## **Example-03:**

```python
class Employee:
  dep_name = 'Python Developer'  # static variable

  def __init__(self, name, age, salary):
    self.name = name
    self.age = age
    self.salary = salary
  
  def Print_Department(self):
    return f"Department Name: {self.dep_name}"
  
emp1 = Employee('Dibyendu', 24, 32000)
print(emp1.Print_Department())

emp2 = Employee('Arko', 29, 45000)
emp2.dep_name = 'Data Analyst'
print(emp2.Print_Department())
```

```python
class Employee:

  dep_name = 'Python Developer'  # static variable

  def __init__(self, name, age, salary):
    self.name = name
    self.age = age
    self.salary = salary
  
  def __str__(self):
    return f"Name: {self.name}, Age: {self.age}, Salary: {self.salary}, Department Name: {self.dep_name}"

emp1 = Employee('Dibyendu', 23, 28000)
print(emp1, '\n')

emp2 = Employee('Ramu', 32, 52000)
emp2.dep_name = "Businee Analyst"
print(emp2, '\n')

emp3 = Employee('Jadav', 44, 75000)
emp3.dep_name = 'Data Science'
print(emp3)
```

# **2. Static Method:**

Python has a static method that belongs to the class. It is just like a static variable that bounds to the class rather than the class's object. A static method can be called without creating an object for the class. It means we can directly call the static method with the reference of the class name. 

**Following are the features of the static method:**

    * A static method in Python related to the class.
    * It can be called directly from the class by reference to a class name.
    * It cannot access the class attributes in the Python program.
    * It is bound only to the class. So it cannot modify the state of the object
    * It is also used to divide the utility methods for the class.
    * It can only be defined inside a class but not to the objects of the class.
    * All the objects of the class share only one copy of the static method.
  

**There are the two ways to define a static method in Python:**

* Using the **staticmethod()** Method,
* Using the **@staticmethod** Decorator.
  > A **@staticmethod** is an inbuilt decorator that defines the static method inside the class. It does not receive any argument as a reference to a class instance or a class calling the static method itself.

  




## **Example-01:**

```python
class Calculations:

  def sum(self, a, b): # define static method or static function
    return a + b
  
  def subtraction(self, a, b): # define static method or static function
    return a - b
  
  def multiplication(self, a, b): # define static method or static function
    return a * b
  
cal1 = Calculations()
print("Addition value: ", cal1.sum(10, 20))
print("Subtraction Value: ", cal1.subtraction(30, 5))
print("Multiplication Value: ", cal1.multiplication(10, 20))
```

## **Example-02:**

```python
class Calculations:

  def __init__(self, a, b):
    self.a = a
    self.b = b

  def sum(self): # define static method or static function
    return self.a + self.b
  
  def subtraction(self): # define static method or static function
    return self.a - self.b
  
  def multiplication(self): # define static method or static function
    return self.a * self.b

  
cal1 = Calculations(30, 2)
print("Addition value: ", cal1.sum())
print("Subtraction Value: ", cal1.subtraction())
print("Multiplication Value: ", cal1.multiplication())
```

## **Example-03:**



```python
class NewCalculations:

  @staticmethod
  def sum(a, b): # define static method or static function
    return a + b
  
  @staticmethod
  def subtraction(a, b): # define static method or static function
    return a - b
  
  @staticmethod
  def multiplication(a, b): # define static method or static function
    return a * b
  
cal1 = NewCalculations()
print("Addition value: ", cal1.sum(10, 20))
print("Subtraction Value: ", cal1.subtraction(30, 5))
print("Multiplication Value: ", cal1.multiplication(10, 20))
```

## **Example-04:**

```python
class Calculations:

  def __init__(self, a, b):
    self.a = a
    self.b = b

  @staticmethod
  def sum(self): # define static method or static function
    return self.a + self.b
  
  @staticmethod
  def subtraction(self): # define static method or static function
    return self.a - self.b
  
  @staticmethod
  def multiplication(self): # define static method or static function
    return self.a * self.b

  
cal1 = Calculations(30, 2)
print("Addition value: ", cal1.sum())
print("Subtraction Value: ", cal1.subtraction())
print("Multiplication Value: ", cal1.multiplication())
```
